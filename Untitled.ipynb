```python
# Test de normalité de Shapiro-Wilk + Transformation Log

# Le test de Shapiro-Wilk teste l'hypothèse nulle :
# H0 : la variable suit une distribution normale
# Si p-value < 0.05 → on rejette H0 → la variable N'EST PAS normale

# Test sur SalePrice brut
stat_raw, p_raw = stats.shapiro(train['SalePrice'].sample(min(5000, len(train)),
                                                           random_state=42))
print(f"Test de Shapiro-Wilk sur SalePrice :")
print(f"  Statistique W = {stat_raw:.6f}")
print(f"  p-value = {p_raw:.6e}")
print(f"   {'NON NORMAL' if p_raw < 0.05 else 'NORMAL'} (p {'<' if p_raw < 0.05 else '>'} 0.05)")

print()

# Transformation logarithmique
# np.log1p(x) = log(x + 1) — le +1 évite log(0) pour les zéros
y_log = np.log1p(train['SalePrice'])

stat_log, p_log = stats.shapiro(y_log.sample(min(5000, len(y_log)), random_state=42))
print(f"Test de Shapiro-Wilk sur log(SalePrice) :")
print(f"  Statistique W = {stat_log:.6f}")
print(f"  p-value = {p_log:.6e}")
print(f"   {'NON NORMAL' if p_log < 0.05 else 'NORMAL'} (p {'<' if p_log < 0.05 else '>'} 0.05)")

print()
print(f"Asymétrie avant log : {train['SalePrice'].skew():.4f}")
print(f"Asymétrie après log : {y_log.skew():.4f}")
print(f" La transformation log RÉDUIT l'asymétrie de {abs(train['SalePrice'].skew()):.2f} à {abs(y_log.skew()):.2f}")
```

```python
# Détection des outliers avec IQR

# Variables numériques les plus importantes à vérifier
# key_vars = ['GrLivArea', 'LotArea', 'TotalBsmtSF', '1stFlrSF',
#             'GarageArea', 'MasVnrArea']
key_vars = ['OverallQual', 'GrLivArea', 'GarageCars','GarageArea', 'TotalBsmtSF',
            'FullBath', 'YearBuilt', 'YearRemodAdd','1stFlrSF','TotRmsAbvGrd']

print("Détection des outliers (méthode IQR, facteur 3.0 pour cas extrêmes) :")

outlier_summary = []
for var in key_vars:
    if var not in train.columns:
        continue
    Q1 = train[var].quantile(0.25)
    Q3 = train[var].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR  # Facteur 3 pour extrêmes seulement
    upper = Q3 + 1.5 * IQR
    n_outliers = ((train[var] < lower) | (train[var] > upper)).sum()
    pct = n_outliers / len(train) * 100
    outlier_summary.append({'Variable': var, 'N_Outliers': n_outliers, 'Pct': pct})
    print(f"  {var:20s} : {n_outliers:4d} outliers extrêmes ({pct:.2f}%)")

print()
print(" On supprimera seulement les cas clairement aberrants documentés par Kaggle :")
print("    GrLivArea > 4000 pieds² ET SalePrice < 300,000 $")
```

```python
#Gestion des valeurs aberrantes par clipping
df_outliers = train.copy()
for col in key_vars:
    Q1 = train[col].quantile(0.25)
    Q3 = train[col].quantile(0.75)

    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR

    train[col] = train[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
def remove_outliers_saleprice(train: pd.DataFrame) -> pd.DataFrame:
    """
    Supprime les outliers identifiés visuellement dans les données d'entraînement.

    D'après la documentation Kaggle du dataset et les recommandations de l'auteur,
    certaines maisons ont une grande surface mais un prix très bas — ce sont des
    ventes atypiques (ex: vente familiale, saisie bancaire) qui perturbent le modèle.

    Paramètres:
      train (DataFrame) : Données d'entraînement avec SalePrice

    Retourne:
      train_clean (DataFrame) : Données sans les outliers
    """
    initial_size = len(train)

    # L'auteur du dataset recommande de supprimer les maisons > 4000 pieds² avec prix < $300,000
    # Ces maisons sont des ventes anormales qui bruitent le modèle
    mask_to_remove = (train['GrLivArea'] > 4000) & (train['SalePrice'] < 300000)

    train_clean = train[~mask_to_remove].copy()  # ~ = "NOT" (inverse le masque)

    removed = initial_size - len(train_clean)
    print(f" Outliers supprimés : {removed} lignes ({removed/initial_size*100:.2f}% des données)")
    print(f"   Taille finale du train : {len(train_clean)} lignes")

    return train_clean
```

```python
#  IMPUTATION DES VALEURS MANQUANTES


def impute_missing_values(df: pd.DataFrame) -> pd.DataFrame:
    """
    Remplace les valeurs manquantes de manière intelligente selon la signification
    de chaque variable dans le contexte immobilier.

    Pour de nombreuses variables (ex: PoolQC, GarageType), NaN signifie simplement
    "cette caractéristique n'existe pas" — ce n'est pas une donnée manquante !

    Paramètres:
      df (DataFrame) : Le tableau de données (train ou test)

    Retourne:
      df (DataFrame) : Le tableau avec toutes les NaN remplacées
    """
    # On travaille sur une copie pour ne pas modifier l'original
    df = df.copy()

    #  VARIABLES CATÉGORIELLES où NaN = "Absent" 
    # Ces variables ont un sens logique : si PoolQC est NaN, c'est qu'il n'y a pas de piscine
    categorical_none = [
        'PoolQC',       # Qualité de la piscine → "No Pool"
        'MiscFeature',  # Caractéristiques diverses → "None"
        'Alley',        # Type d'allée → "No Alley"
        'Fence',        # Clôture → "No Fence"
        'FireplaceQu',  # Qualité de la cheminée → "No Fireplace"
        'GarageType',   # Type de garage → "No Garage"
        'GarageFinish', # Finition du garage → "No Garage"
        'GarageQual',   # Qualité du garage → "No Garage"
        'GarageCond',   # Condition du garage → "No Garage"
        'BsmtQual',     # Qualité du sous-sol → "No Basement"
        'BsmtCond',     # Condition du sous-sol → "No Basement"
        'BsmtExposure', # Exposition du sous-sol → "No Basement"
        'BsmtFinType1', # Type finition sous-sol 1 → "No Basement"
        'BsmtFinType2', # Type finition sous-sol 2 → "No Basement"
        'MasVnrType',   # Type de parement en maçonnerie → "None"
    ]

    for col in categorical_none:
        if col in df.columns:
            # fillna() remplace les NaN par la valeur spécifiée
            df[col] = df[col].fillna('None')

    # --- VARIABLES NUMÉRIQUES où NaN = 0 (logique : absent = 0 m² ou 0 voiture) ---
    numerical_zero = [
        'GarageYrBlt',  # Année construction garage → 0 si pas de garage
        'GarageArea',   # Surface du garage → 0
        'GarageCars',   # Capacité du garage → 0
        'BsmtFinSF1',   # Surface finie sous-sol 1 → 0
        'BsmtFinSF2',   # Surface finie sous-sol 2 → 0
        'BsmtUnfSF',    # Surface non-finie sous-sol → 0
        'TotalBsmtSF',  # Surface totale sous-sol → 0
        'BsmtFullBath', # Salles de bain complètes sous-sol → 0
        'BsmtHalfBath', # Demi salles de bain sous-sol → 0
        'MasVnrArea',   # Surface parement maçonnerie → 0
    ]

    for col in numerical_zero:
        if col in df.columns:
            df[col] = df[col].fillna(0)

    #  LotFrontage : distance de la rue — on utilise la MÉDIANE par quartier 
    # La médiane est plus robuste que la moyenne car elle ignore les valeurs extrêmes
    # Logique : les maisons du même quartier ont des rues similaires
    if 'LotFrontage' in df.columns and 'Neighborhood' in df.columns:
        df['LotFrontage'] = df.groupby('Neighborhood')['LotFrontage'].transform(
            lambda x: x.fillna(x.median())
        )
        # Si encore des NaN (quartier avec toutes les valeurs manquantes), on met la médiane globale
        df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())

    #  VARIABLES CATÉGORIELLES restantes → MODE (valeur la plus fréquente) 
    remaining_categorical = df.select_dtypes(include=['object']).columns
    for col in remaining_categorical:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna(df[col].mode()[0])

    #  VARIABLES NUMÉRIQUES restantes -> MÉDIANE 
    remaining_numerical = df.select_dtypes(include=[np.number]).columns
    for col in remaining_numerical:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna(df[col].median())

    # Vérification finale
    remaining_na = df.isnull().sum().sum()
    print(f" Imputation terminée. Valeurs manquantes restantes : {remaining_na}")

    return df
```

```python
# 1. ENCODAGE DES VARIABLES ORDINALES
# ==============================================================

def encode_ordinal_variables(df: pd.DataFrame) -> pd.DataFrame:
    """
    Encode les variables ordinales (qui ont un ordre naturel).

    Une variable ordinale est une variable catégorielle dont les valeurs
    ont un sens de classement. Exemple :
      'Po' (Poor) < 'Fa' (Fair) < 'TA' (Average) < 'Gd' (Good) < 'Ex' (Excellent)

    On les transforme en nombres : Po=1, Fa=2, TA=3, Gd=4, Ex=5

    IMPORTANT : Il ne faut PAS utiliser One-Hot Encoding pour ces variables,
    car cela perdrait l'information d'ordre (Ex > Gd > TA).

    Paramètres:
      df (DataFrame) : Le tableau de données

    Retourne:
      df (DataFrame) : Tableau avec les variables ordinales encodées en nombres
    """
    df = df.copy()

    # Mapping qualité générale : Poor -> 1 jusqu'à Excellent -> 5
    quality_map = {'None': 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5}

    # Liste des colonnes de qualité qui utilisent le même barème
    quality_cols = [
        'ExterQual', 'ExterCond',    # Qualité/condition extérieure
        'BsmtQual', 'BsmtCond',      # Qualité/condition du sous-sol
        'HeatingQC',                  # Qualité du chauffage
        'KitchenQual',               # Qualité de la cuisine
        'FireplaceQu',               # Qualité de la cheminée
        'GarageQual', 'GarageCond',  # Qualité/condition du garage
        'PoolQC',                    # Qualité de la piscine
    ]

    for col in quality_cols:
        if col in df.columns:
            # map() remplace chaque valeur selon le dictionnaire
            df[col] = df[col].map(quality_map)
            # Remplace les valeurs non trouvées dans le dictionnaire par 0
            df[col] = df[col].fillna(0)

    #  Exposition du sous-sol 
    bsmt_exposure_map = {'None': 0, 'No': 1, 'Mn': 2, 'Av': 3, 'Gd': 4}
    if 'BsmtExposure' in df.columns:
        df['BsmtExposure'] = df['BsmtExposure'].map(bsmt_exposure_map).fillna(0)

    #  Type de finition du sous-sol 
    bsmt_fin_map = {'None': 0, 'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ': 6}
    for col in ['BsmtFinType1', 'BsmtFinType2']:
        if col in df.columns:
            df[col] = df[col].map(bsmt_fin_map).fillna(0)

    #  Finition du garage 
    garage_fin_map = {'None': 0, 'Unf': 1, 'RFn': 2, 'Fin': 3}
    if 'GarageFinish' in df.columns:
        df['GarageFinish'] = df['GarageFinish'].map(garage_fin_map).fillna(0)

    #  Allée 
    alley_map = {'None': 0, 'Grvl': 1, 'Pave': 2}
    if 'Alley' in df.columns:
        df['Alley'] = df['Alley'].map(alley_map).fillna(0)

    #  Pente du terrain 
    slope_map = {'Gtl': 1, 'Mod': 2, 'Sev': 3}
    if 'LandSlope' in df.columns:
        df['LandSlope'] = df['LandSlope'].map(slope_map).fillna(1)

    #  Fonctionnalité de la maison 
    functional_map = {'Sal': 1, 'Sev': 2, 'Maj2': 3, 'Maj1': 4,
                      'Mod': 5, 'Min2': 6, 'Min1': 7, 'Typ': 8}
    if 'Functional' in df.columns:
        df['Functional'] = df['Functional'].map(functional_map).fillna(7)

    #  Clôture 
    fence_map = {'None': 0, 'MnWw': 1, 'GdWo': 2, 'MnPrv': 3, 'GdPrv': 4}
    if 'Fence' in df.columns:
        df['Fence'] = df['Fence'].map(fence_map).fillna(0)

    #  Allée pavée 
    paved_map = {'N': 0, 'P': 1, 'Y': 2}
    if 'PavedDrive' in df.columns:
        df['PavedDrive'] = df['PavedDrive'].map(paved_map).fillna(0)

    #  Air conditionné 
    if 'CentralAir' in df.columns:
        df['CentralAir'] = df['CentralAir'].map({'N': 0, 'Y': 1}).fillna(0)

    print(" Variables ordinales encodées")
    return df
```

```python
#  CRÉATION DE NOUVELLES FEATURES (FEATURE ENGINEERING)

def create_new_features(df: pd.DataFrame) -> pd.DataFrame:
    """
    Crée au minimum 5 nouvelles variables pertinentes pour la prédiction du prix.

    Chaque nouvelle variable capture une information que le modèle ne peut pas
    facilement déduire seul à partir des variables originales.

    Paramètres:
      df (DataFrame) : Le tableau de données

    Retourne:
      df (DataFrame) : Tableau enrichi avec les nouvelles features
    """
    df = df.copy()

    # -------------------------------------------------------
    # FEATURE 1 : Surface Totale de la Maison (TotalSF)
    # -------------------------------------------------------
    # La surface totale est le facteur N°1 de valeur d'une maison.
    # On combine : sous-sol + rez-de-chaussée + étage
    df['TotalSF'] = (
        df.get('TotalBsmtSF', 0) +   # Surface totale du sous-sol
        df.get('1stFlrSF', 0) +       # Surface rez-de-chaussée
        df.get('2ndFlrSF', 0)         # Surface du 1er étage (américain)
    )
    print("   Feature 1/8 : TotalSF (surface totale)")

    # -------------------------------------------------------
    # FEATURE 2 : Âge de la Maison (HouseAge)
    # -------------------------------------------------------
    # Une maison récente vaut généralement plus qu'une ancienne.
    # On calcule l'âge au moment de la vente.
    if 'YrSold' in df.columns and 'YearBuilt' in df.columns:
        df['HouseAge'] = df['YrSold'] - df['YearBuilt']
        # Valeurs négatives impossibles → on met 0
        df['HouseAge'] = df['HouseAge'].clip(lower=0)
        print("   Feature 2/8 : HouseAge (âge de la maison lors de la vente)")

    # -------------------------------------------------------
    # FEATURE 3 : Années depuis la dernière rénovation (YearsSinceRemod)
    # -------------------------------------------------------
    # Une maison rénovée récemment vaut plus qu'une maison non rénovée.
    if 'YrSold' in df.columns and 'YearRemodAdd' in df.columns:
        df['YearsSinceRemod'] = df['YrSold'] - df['YearRemodAdd']
        df['YearsSinceRemod'] = df['YearsSinceRemod'].clip(lower=0)
        print("   Feature 3/8 : YearsSinceRemod (années depuis dernière rénovation)")

    # -------------------------------------------------------
    # FEATURE 4 : Nombre total de salles de bain (TotalBathrooms)
    # -------------------------------------------------------
    # Les acheteurs regardent souvent le nombre de salles de bain total.
    # Une demi-salle de bain (juste lavabo + toilettes) compte pour 0.5
    df['TotalBathrooms'] = (
        df.get('FullBath', 0) +           # SdB complètes au-dessus du sol
        df.get('HalfBath', 0) * 0.5 +     # Demi-SdB au-dessus du sol
        df.get('BsmtFullBath', 0) +        # SdB complètes sous-sol
        df.get('BsmtHalfBath', 0) * 0.5   # Demi-SdB sous-sol
    )
    print("   Feature 4/8 : TotalBathrooms (total salles de bain)")

    # -------------------------------------------------------
    # FEATURE 5 : Surface totale des vérandas/porches (TotalPorchSF)
    # -------------------------------------------------------
    # Les espaces extérieurs ajoutent de la valeur
    df['TotalPorchSF'] = (
        df.get('OpenPorchSF', 0) +     # Porche ouvert
        df.get('EnclosedPorch', 0) +   # Porche fermé
        df.get('3SsnPorch', 0) +       # Porche 3 saisons
        df.get('ScreenPorch', 0) +     # Porche grillagé
        df.get('WoodDeckSF', 0)        # Terrasse en bois
    )
    print("   Feature 5/8 : TotalPorchSF (surface totale extérieure)")

    # -------------------------------------------------------
    # FEATURE 6 : La maison a-t-elle été rénovée ? (IsRemodeled)
    # -------------------------------------------------------
    # Variable binaire (0 ou 1) : 1 si rénovée, 0 sinon
    if 'YearBuilt' in df.columns and 'YearRemodAdd' in df.columns:
        df['IsRemodeled'] = (df['YearBuilt'] != df['YearRemodAdd']).astype(int)
        print("   Feature 6/8 : IsRemodeled (maison rénovée ou non)")

    # -------------------------------------------------------
    # FEATURE 7 : La maison est-elle neuve ? (IsNewHouse)
    # -------------------------------------------------------
    # Les maisons neuves (construites et vendues la même année) ont un prix premium
    if 'YrSold' in df.columns and 'YearBuilt' in df.columns:
        df['IsNewHouse'] = (df['YrSold'] == df['YearBuilt']).astype(int)
        print("   Feature 7/8 : IsNewHouse (maison neuve = construite et vendue la même année)")

    # -------------------------------------------------------
    # FEATURE 8 : Score composite qualité × surface (QualArea)
    # -------------------------------------------------------
    # L'interaction entre qualité et surface est très prédictive du prix.
    # Une grande maison de bonne qualité vaut beaucoup plus qu'une grande maison de mauvaise qualité.
    if 'OverallQual' in df.columns:
        df['QualArea'] = df['OverallQual'] * df['TotalSF']
        print("   Feature 8/8 : QualArea (qualité × surface totale)")

    print(f"\n Nouvelles features créées. Dimensions : {df.shape}")
    return df
```

```python
#  SUPPRESSION DES COLONNES INUTILES

def drop_useless_columns(df: pd.DataFrame) -> pd.DataFrame:
    """
    Supprime les colonnes qui n'apportent pas d'information utile au modèle.

    Paramètres:
      df (DataFrame) : Le tableau de données

    Retourne:
      df (DataFrame) : Tableau sans les colonnes inutiles
    """
    cols_to_drop = [
        'Id',           # Numéro d'identifiant — aucune valeur prédictive
        'Utilities',    # Quasiment toutes les maisons ont les mêmes utilities → pas informatif
        'PoolQC',       # Très peu de maisons ont une piscine → beaucoup de 'None'
        'MiscFeature',  # Rare et peu impactant
        'MiscVal',      # Lié à MiscFeature, souvent 0
        'Alley',               
        'Fence',                 
        'MasVnrType',             
        'FireplaceQu',
    ]

    # On ne supprime que les colonnes qui existent réellement dans le DataFrame
    cols_to_drop = [c for c in cols_to_drop if c in df.columns]

    df = df.drop(columns=cols_to_drop)
    print(f" Colonnes supprimées : {cols_to_drop}")

    return df
```

```python
# 3. ENCODAGE ONE-HOT DES VARIABLES NOMINALES

def encode_categorical_variables(train: pd.DataFrame, test: pd.DataFrame):
    """
    Encode les variables catégorielles nominales avec One-Hot Encoding.

    Le One-Hot Encoding transforme une variable catégorielle en plusieurs
    colonnes binaires (0 ou 1). Exemple :
      Neighborhood: 'CollgCr' → [1,0,0,...,0]
      Neighborhood: 'OldTown' → [0,1,0,...,0]

    On ne peut PAS utiliser les nombres (1, 2, 3...) pour les variables nominales
    car cela créerait un ordre artificiel inexistant.

    Paramètres:
      train (DataFrame) : Données d'entraînement
      test  (DataFrame) : Données de test

    Retourne:
      train_enc (DataFrame), test_enc (DataFrame) : Données encodées
    """
    # On identifie les colonnes encore catégorielles (texte) après l'encodage ordinal
    categorical_cols = train.select_dtypes(include=['object']).columns.tolist()

    print(f" Variables nominales à encoder ({len(categorical_cols)}) : {categorical_cols}")

    # pd.get_dummies() fait le One-Hot Encoding
    # drop_first=True : on supprime la première modalité pour éviter la multicolinéarité
    # (avec k modalités, on n'a besoin que de k-1 colonnes binaires)
    train_enc = pd.get_dummies(train, columns=categorical_cols, drop_first=True)
    test_enc  = pd.get_dummies(test,  columns=categorical_cols, drop_first=True)

    # IMPORTANT : Aligner les colonnes de train et test
    # Le test peut avoir des modalités différentes → on doit avoir les mêmes colonnes
    train_enc, test_enc = train_enc.align(test_enc, join='left', axis=1, fill_value=0)

    print(f" Encodage terminé.")
    print(f"   Train : {train_enc.shape[1]} colonnes")
    print(f"   Test  : {test_enc.shape[1]} colonnes")

    return train_enc, test_enc

```

```python
# 5. TRANSFORMATION DE LA VARIABLE CIBLE
# ==============================================================

def log_transform_target(y: pd.Series) -> pd.Series:
    """
    Applique une transformation logarithmique à la variable cible SalePrice.

    Pourquoi ?
    - SalePrice a une distribution asymétrique à droite (quelques maisons
      très chères tirent la moyenne vers le haut)
    - Les algorithmes de régression fonctionnent mieux avec une distribution
      proche de la loi normale (en cloche)
    - La transformation log(SalePrice) rend la distribution plus symétrique
    - En plus, l'erreur RMSE sur log(price) = RMSLE, ce qui pénalise moins
      les erreurs relatives sur les maisons chères

    Paramètres:
      y (Series) : La colonne SalePrice

    Retourne:
      y_log (Series) : log(SalePrice + 1) — le +1 évite log(0)
    """
    y_log = np.log1p(y)  # log1p(x) = log(x + 1) — plus stable que log(x)
    print(f" Transformation logarithmique appliquée sur SalePrice")
    print(f"   Avant : moyenne={y.mean():.0f}, asymétrie={y.skew():.2f}")
    print(f"   Après : moyenne={y_log.mean():.4f}, asymétrie={y_log.skew():.2f}")
    return y_log
```

```python
#  SCALING (NORMALISATION DES FEATURES)


def scale_features(X_train: pd.DataFrame, X_test: pd.DataFrame):
    """
    Normalise les features numériques avec StandardScaler.

    Le StandardScaler transforme chaque variable pour qu'elle ait :
      - Moyenne = 0
      - Écart-type = 1

    Pourquoi normaliser ?
    - Certains modèles (Ridge, Lasso, ElasticNet) sont sensibles à l'échelle
      des variables. Une variable en m² (0-500) dominera une variable 0/1 si
      on ne normalise pas.
    - Les arbres (Random Forest, Gradient Boosting) n'en ont pas besoin,
      mais ça ne fait pas de mal.

    IMPORTANT : On ajuste le scaler UNIQUEMENT sur les données d'entraînement,
    puis on l'applique au test. Sinon, il y aurait une "fuite" d'information
    du test vers l'entraînement (data leakage).

    Paramètres:
      X_train (DataFrame) : Features d'entraînement
      X_test  (DataFrame) : Features de test

    Retourne:
      X_train_scaled, X_test_scaled (arrays numpy), scaler (objet StandardScaler)
    """
    scaler = StandardScaler()

    # fit() calcule la moyenne et l'écart-type sur les données d'entraînement
    # transform() applique la normalisation
    X_train_scaled = scaler.fit_transform(X_train)

    # transform() seulement (pas fit_transform) — on utilise les stats du train
    X_test_scaled = scaler.transform(X_test)

    print(" Scaling appliqué (StandardScaler)")
    print(f"   Moyennes après scaling : ~{X_train_scaled.mean():.4f} (devrait être ≈ 0)")
    print(f"   Écart-types après scaling : ~{X_train_scaled.std():.4f} (devrait être ≈ 1)")

    return X_train_scaled, X_test_scaled, scaler

```

```python
# Définition de tous les modèles

# Configuration de la validation croisée
# shuffle=True : mélange les données avant de créer les folds
# random_state=42 : graine pour la reproductibilité
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# ---- MODÈLES LINÉAIRES (nécessitent X_train_scaled) ----

# Régression Linéaire classique — aucune régularisation
linear_reg = LinearRegression()

# Ridge : alpha = force de la régularisation
# Grand alpha = plus de régularisation = coefficients plus petits
ridge = Ridge(alpha=10.0, random_state=42 if hasattr(Ridge, 'random_state') else None)
ridge = Ridge(alpha=10.0)

# Lasso : alpha = même principe
# selection='cyclic' : algorithme d'optimisation
lasso = Lasso(alpha=0.0005, max_iter=5000, random_state=42)

# ElasticNet : l1_ratio = proportion L1 (0=Ridge pur, 1=Lasso pur)
elastic = ElasticNet(alpha=0.0005, l1_ratio=0.9, max_iter=5000, random_state=42)

# ---- MODÈLES À BASE D'ARBRES (pas besoin de scaling) ----

# Random Forest : n_estimators = nombre d'arbres
# max_depth = profondeur max des arbres (None = pas de limite)
rf = RandomForestRegressor(
    n_estimators=500,     # 500 arbres → bon équilibre précision/vitesse
    max_features='sqrt',  # Nb features testées à chaque split = sqrt(n_features)
    max_depth=None,       # Arbres profonds (complets)
    min_samples_leaf=2,   # Chaque feuille doit contenir au moins 2 exemples
    random_state=42,
    n_jobs=-1             # Utilise tous les cœurs CPU
)

# Gradient Boosting — version scikit-learn
gb = GradientBoostingRegressor(
    n_estimators=1000,    # 1000 arbres ajoutés séquentiellement
    learning_rate=0.05,   # Taux d'apprentissage (petit = meilleur mais lent)
    max_depth=4,          # Profondeur des arbres individuels
    min_samples_leaf=15,  # Régularisation
    subsample=0.8,        # 80% des données par itération (stochastic GB)
    random_state=42
)

# Dictionnaire modèle_nom → (modèle, utilise_scaling)
# True = on utilise X_train_scaled (modèles linéaires)
# False = on utilise X_train brut (arbres)
models = {
    'Linear Regression':   (linear_reg, True),
    'Ridge':               (ridge,      True),
    'Lasso':               (lasso,      True),
    'ElasticNet':          (elastic,    True),
    'Random Forest':       (rf,         False),
    'Gradient Boosting':   (gb,         False),
}

# Ajout XGBoost si disponible
if XGBOOST_AVAILABLE:
    xgb = XGBRegressor(
        n_estimators=2000,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.01,
        reg_lambda=0.01,
        random_state=42,
        n_jobs=-1,
        verbosity=0
    )
    models['XGBoost'] = (xgb, False)

print(f" {len(models)} modèles configurés :")
for name in models:
    print(f"   - {name}")
```

```python
#  4 : Validation croisée de tous les modèles

# C'est la cellule principale — elle peut prendre quelques minutes

results = []

print("Entraînement et évaluation par validation croisée 5-Fold...")
print(f"{'Modèle':25s} | {'RMSE Moy':10s} | {'RMSE Std':10s} | {'R² Moy':8s} | {'R² Std':8s}")


for model_name, (model, needs_scaling) in models.items():
    # Choix des données : scalées pour modèles linéaires, brutes pour arbres
    X_to_use = X_train_scaled if needs_scaling else X_train
    
    # cross_val_score évalue le modèle avec K-Fold
    # neg_mean_squared_error : MSE négatif (sklearn maximise → on minimise MSE)
    neg_mse = cross_val_score(
        model, X_to_use, y_log,
        cv=kf,
        scoring='neg_mean_squared_error',
        n_jobs=-1
    )
    # Conversion en RMSE positif
    rmse_scores = np.sqrt(-neg_mse)
    
    r2_scores = cross_val_score(
        model, X_to_use, y_log,
        cv=kf,
        scoring='r2',
        n_jobs=-1
    )
    
    result = {
        'model_name': model_name,
        'model': model,
        'needs_scaling': needs_scaling,
        'rmse_mean': rmse_scores.mean(),
        'rmse_std': rmse_scores.std(),
        'r2_mean': r2_scores.mean(),
        'r2_std': r2_scores.std(),
        'rmse_scores': rmse_scores,
        'r2_scores': r2_scores
    }
    results.append(result)
    
    print(f"{model_name:25s} | {rmse_scores.mean():.5f}    | {rmse_scores.std():.5f}    | "
          f"{r2_scores.mean():.4f}   | {r2_scores.std():.4f}")

print()
print(" Validation croisée terminée !")
```

```python

#  7 : Analyse Biais / Variance
# On analyse le Gap entre performance ENTRAÎNEMENT et VALIDATION
# pour chaque modèle.
#
# Biais élevé (underfitting) : erreur train ET val sont élevées
# Variance élevée (overfitting) : erreur train << val (grand gap)

print("Analyse Biais / Variance :")
print(f"{'Modèle':25s} | {'RMSE Train':12s} | {'RMSE Val':12s} | {'Gap':10s} | {'Diagnostic'}")

# On utilise un split simple pour l'analyse biais/variance
from sklearn.model_selection import train_test_split

X_split_train, X_split_val, y_split_train, y_split_val = train_test_split(
    X_train, y_log, test_size=0.2, random_state=42
)
X_split_train_sc, X_split_val_sc, _ = scale_features(
    pd.DataFrame(X_split_train), pd.DataFrame(X_split_val)
)

for r in results:
    model = r['model']
    if r['needs_scaling']:
        X_tr, X_vl = X_split_train_sc, X_split_val_sc
    else:
        X_tr, X_vl = X_split_train, X_split_val
    
    model.fit(X_tr, y_split_train)
    
    rmse_train = np.sqrt(mean_squared_error(y_split_train, model.predict(X_tr)))
    rmse_val   = np.sqrt(mean_squared_error(y_split_val,   model.predict(X_vl)))
    gap = rmse_val - rmse_train
    
    if gap > 0.04:
        diag = "  Overfitting"
    elif rmse_train > 0.20:
        diag = "  Underfitting"
    else:
        diag = " Bon équilibre"
    
    print(f"{r['model_name']:25s} | {rmse_train:.5f}      | {rmse_val:.5f}      | {gap:.5f}    | {diag}")
```

```python

#  8 : Courbe d'apprentissage du meilleur modèle

from sklearn.model_selection import learning_curve

# On prend le meilleur modèle (GradientBoosting ou XGBoost)
best_model_result = min(results, key=lambda x: x['rmse_mean'])
best_model = best_model_result['model']
best_name  = best_model_result['model_name']
X_best = X_train_scaled if best_model_result['needs_scaling'] else X_train

print(f"Calcul des courbes d'apprentissage pour : {best_name}")
print("(Peut prendre 2-3 minutes...)")

train_sizes_pct = np.linspace(0.1, 1.0, 8)

train_sizes_abs, train_scores, val_scores = learning_curve(
    best_model,
    X_best, y_log,
    train_sizes=train_sizes_pct,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1
)

train_rmse = np.sqrt(-train_scores.mean(axis=1))
val_rmse   = np.sqrt(-val_scores.mean(axis=1))
train_std  = np.sqrt(-train_scores).std(axis=1)
val_std    = np.sqrt(-val_scores).std(axis=1)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(train_sizes_abs, train_rmse, 'b-o', label='Erreur Entraînement', linewidth=2.5)
ax.fill_between(train_sizes_abs, train_rmse - train_std, train_rmse + train_std,
                alpha=0.2, color='blue')

ax.plot(train_sizes_abs, val_rmse, 'r-o', label='Erreur Validation', linewidth=2.5)
ax.fill_between(train_sizes_abs, val_rmse - val_std, val_rmse + val_std,
                alpha=0.2, color='red')

# Annotation de la zone de convergence
gap_final = val_rmse[-1] - train_rmse[-1]
ax.annotate(f'Gap final = {gap_final:.4f}',
            xy=(train_sizes_abs[-1], (val_rmse[-1] + train_rmse[-1])/2),
            xytext=(train_sizes_abs[-2], (val_rmse[-1] + train_rmse[-1])/2 + 0.005),
            arrowprops=dict(arrowstyle='->', color='black'),
            fontsize=11, color='black')

ax.set_xlabel("Nombre d'Exemples d'Entraînement", fontsize=12)
ax.set_ylabel('RMSE', fontsize=12)
ax.set_title(f'Courbes d\'Apprentissage — {best_name}', fontsize=13, fontweight='bold')
ax.legend(fontsize=12)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../modelsNouv/14_learning_curve.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\n INTERPRÉTATION :")
print(f"   Gap train/val final = {gap_final:.4f}")
if gap_final > 0.04:
    print("     Overfitting : le modèle mémorise les données d'entraînement")
    print("    Solution : régularisation plus forte, moins d'arbres, early stopping")
elif gap_final < 0.005:
    print("     Underfitting : le modèle est trop simple")
    print("    Solution : plus d'arbres, moins de régularisation")
else:
    print("    Bon équilibre biais/variance : le modèle généralise bien")
```

```python
#  9 : Analyse des résidus du meilleur modèle

from sklearn.model_selection import train_test_split as tts

# Split 80/20
X_tr, X_vl, y_tr, y_vl = tts(
    X_best if not best_model_result['needs_scaling'] else X_train_scaled,
    y_log, test_size=0.2, random_state=42
)

best_model.fit(X_tr, y_tr)
y_pred = best_model.predict(X_vl)
residuals = y_vl - y_pred

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(f'Analyse des Résidus — {best_name}', fontsize=13, fontweight='bold')

# Graphique 1 : Résidus vs Prédictions
axes[0].scatter(y_pred, residuals, alpha=0.4, color='steelblue', s=20, edgecolors='none')
axes[0].axhline(y=0, color='red', linestyle='--', linewidth=2)
axes[0].set_xlabel('Valeurs Prédites (log-espace)')
axes[0].set_ylabel('Résidus')
axes[0].set_title('Prédictions vs Résidus')

# Graphique 2 : Distribution des résidus
axes[1].hist(residuals, bins=50, color='steelblue', edgecolor='white', alpha=0.8)
axes[1].axvline(x=0, color='red', linestyle='--', linewidth=2)
axes[1].set_xlabel('Résidus')
axes[1].set_ylabel('Fréquence')
axes[1].set_title(f'Distribution (asymétrie={residuals.mean():.4f})')

# Graphique 3 : Réels vs Prédits
y_real = np.expm1(y_vl)
y_pred_real = np.expm1(y_pred)
axes[2].scatter(y_real, y_pred_real, alpha=0.4, color='#27ae60', s=20, edgecolors='none')
min_v = min(y_real.min(), y_pred_real.min())
max_v = max(y_real.max(), y_pred_real.max())
axes[2].plot([min_v, max_v], [min_v, max_v], 'r--', linewidth=2, label='Parfait')
axes[2].set_xlabel('Prix Réels ($)')
axes[2].set_ylabel('Prix Prédits ($)')
axes[2].set_title('Réels vs Prédits')
axes[2].legend()
axes[2].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))
axes[2].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

plt.tight_layout()
plt.savefig('../modelsNouv/15_residuals.png', dpi=150, bbox_inches='tight')
plt.show()

rmse_final = np.sqrt(mean_squared_error(y_vl, y_pred))
r2_final = r2_score(y_vl, y_pred)
print(f"RMSE final (log-espace) : {rmse_final:.5f}")
print(f"R²   final              : {r2_final:.4f}")
print(f"RMSE en dollars         : ${np.sqrt(mean_squared_error(y_real, y_pred_real)):,.0f}")
```

```python

#  4 : Optimisation GradientBoosting avec RandomizedSearchCV

# RandomizedSearchCV est preferable quand l'espace de recherche est grand.
# On tire aleatoirement n_iter combinaisons.
# Avantage : beaucoup plus rapide que GridSearch exhaustif.

print("Optimisation GradientBoosting avec RandomizedSearchCV...")
print("(Peut prendre 10-20 minutes selon votre machine)")

gb_param_dist = {
    'n_estimators': [500, 1000, 1500, 2000, 3000],
    'learning_rate': [0.01, 0.03, 0.05, 0.1],
    'max_depth': [3, 4, 5, 6],
    'min_samples_leaf': [5, 10, 15, 20],
    'subsample': [0.7, 0.8, 0.9, 1.0],
    'max_features': ['sqrt', 'log2'],
}

gb_rs = RandomizedSearchCV(
    estimator=GradientBoostingRegressor(random_state=42),
    param_distributions=gb_param_dist,
    n_iter=30,        # On teste 30 combinaisons aleatoires
    cv=KFold(n_splits=5, shuffle=True, random_state=42),
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1,
    random_state=42,
    refit=True
)

gb_rs.fit(X_train, y_log)

print()
print("Meilleurs hyperparametres trouves :")
for param, value in gb_rs.best_params_.items():
    print(f"   {param:20s} = {value}")

best_gb_score = np.sqrt(-gb_rs.best_score_)
print(f"Meilleur RMSE (CV) : {best_gb_score:.5f}")

# Sauvegarde du meilleur modele
os.makedirs('../modelsNouv', exist_ok=True)
joblib.dump(gb_rs.best_estimator_, '../modelsNouv/best_model.pkl')
print("Meilleur modele sauvegarde dans ../modelsNouv/best_model.pkl")
```
